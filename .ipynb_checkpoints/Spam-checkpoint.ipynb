```python
import pandas as pd
import plotly.graph_objects as go
from sklearn.preprocessing  import LabelEncoder
from sklearn.linear_model import LinearRegression ,LogisticRegression
```

```python
dataset=pd.read_csv("SPAM text message 20170820 - Data.csv")
```

```python
dataset.head(3)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
fig=go.Figure(data=go.Histogram(x=dataset["Category"]))
fig.show()
```

```python
La=LabelEncoder()
dataset["Category_Encoded"]=La.fit_transform(dataset["Category"])
```

```python
dataset["Message_Encoded"]=La.fit_transform(dataset["Message"])
```

```python
dataset.head(3)
```

```python
lr=LogisticRegression()
lr.fit(dataset[["Category_Encoded"]],dataset[["Message_Encoded"]])
```

```python
text=input("")
```

```python
text_en=La.fit_transform([[text]])
```

```python
predicted=pre=lr.predict([text_en])
```

```python
if predicted==1:
    print("This message is not spam")
else:
    print("This message is  spam")
```
